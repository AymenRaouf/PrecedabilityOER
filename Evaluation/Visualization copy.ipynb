```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

main_publisher = "Stanford" # MIT, OYC, Khan, Stanford
chosen_models = ['Conv', 'GAT', 'Reinjection', 'SAGE']
models_labels = []
accuracy = []
accuracy_std = []
time = []
time_std = []
```

```python
methods = "embedd-er" # BERT , embedd-er
class_type = "rdfs" # rdfs, dct

df_ModelsAll = pd.read_csv(f'../Output/Ablation/results_{main_publisher}_{methods}_{class_type}.csv')

accuracy.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Test Accuracy"].values.mean()*100,2))
accuracy_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Test Accuracy"].values.std()*100,2))
time.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Time"].values.mean(), 2))
time_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Time"].values.std(), 2))

df_ModelsAll.head()
```

```python
methods = "BERT" # BERT , embedd-er
class_type = "dct" # rdfs, dct

df_ModelsAll = pd.read_csv(f'../Output/Ablation/results_{main_publisher}_{methods}_{class_type}.csv')

accuracy.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Test Accuracy"].values.mean()*100,2))
accuracy_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Test Accuracy"].values.std()*100,2))
time.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Time"].values.mean(), 2))
time_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == "SAGE"]["Time"].values.std(), 2))

df_ModelsAll[df_ModelsAll["Model"] == "SAGE"].head()
```

```python
methods = "embedd-er" # BERT , embedd-er
class_type = "dct" # rdfs, dct
df_ModelsAll = pd.read_csv(f'../Output/Ablation/results_{main_publisher}_{methods}_{class_type}.csv')
df_ModelsAll.head()
```

```python
for chosen_model in chosen_models:
    accuracy.append(round(df_ModelsAll[df_ModelsAll["Model"] == chosen_model]["Test Accuracy"].values.mean()*100,2))
    accuracy_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == chosen_model]["Test Accuracy"].values.std()*100,2))
    time.append(round(df_ModelsAll[df_ModelsAll["Model"] == chosen_model]["Time"].values.mean(), 2))
    time_std.append(round(df_ModelsAll[df_ModelsAll["Model"] == chosen_model]["Time"].values.std(), 2))
```

```python
all_models = ["RDFT", "BERT"]
all_models += chosen_models
model_labels = ["RDFT", "BERT", "Conv", "GAT", "Rein", "SAGE"]
```

```python
len(all_models)
```

```python
x = np.arange(len(model_labels))


def autolabel(rects, n, first, second):
    for rect in rects:
        height = rect.get_height()
        if height == first:
            axs[n].annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight = 'bold',
                    rotation = 45, size = 10)
        elif height == second:
            axs[n].annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    rotation = 45, size = 10)
        else :
            axs[n].annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight = 'light',
                    rotation = 45, size = 10)


# Plotting the bar plots
width = 0.6  # width of the bars
fig, axs = plt.subplots(1, 2, figsize=(6, 1.5))

first_accuracy = np.sort(np.unique(accuracy))[::-1][0]
second_accuracy = np.sort(np.unique(accuracy))[::-1][1]

first_time = np.sort(np.unique(time))[::-1][-1]
second_time = np.sort(np.unique(time))[::-1][-2]

rects_accuracy = []
rects_accuracy.append(axs[0].bar(x, accuracy, width, yerr = accuracy_std, label = 'Accuracy',color = 'lightsteelblue', 
                                 edgecolor = 'grey', capsize = 8, ecolor = 'saddlebrown'))
#rects_accuracy[0][-1].set_hatch('///')
autolabel(rects_accuracy[0], 0, first_accuracy, second_accuracy)
axs[0].set_title('Accuracy (%)', pad=25, size=10)
axs[0].set_xticks(x)
axs[0].set_xticklabels(model_labels, rotation=45)
axs[0].spines['top'].set_visible(False)
axs[0].spines['right'].set_visible(False)
axs[0].title.set_position([0.5, .05])
#axs[0].legend([rects_accuracy[0][-1], rects_accuracy[0][0]],['Original model', 'Derived models'])

rects_time = []
rects_time.append(axs[1].bar(x, time, width, yerr = time_std, label = 'Time', color = 'peachpuff', 
                                 edgecolor = 'grey', capsize = 8, ecolor = 'saddlebrown'))

#rects_time[0][-1].set_hatch('///')
autolabel(rects_time[0], 1, first_time, second_time)
axs[1].set_title('Time (s)', pad=25, size=10)
axs[1].set_xticks(x)
axs[1].set_xticklabels(model_labels, rotation=45)
axs[1].spines['top'].set_visible(False)
axs[1].spines['right'].set_visible(False)
axs[1].title.set_position([0.5, .05])
title = main_publisher if main_publisher != "OYC" else "Yale"
fig.suptitle(title, fontsize = 16, y = 1.4)
plt.show()
```
