```python
import os
import pandas as pd

main_publisher = 'OYC'
script_dir = os.path.dirname(os.path.realpath('__file__'))
path = os.path.join(script_dir, '../Data/'+ main_publisher + '/')
df_main_series = pd.read_csv(path + 'series.csv', sep = '|')
df_main_episodes = pd.read_csv(path + 'episodes.csv', sep = '|')
df_main_chapters = pd.read_csv(path + 'chapters.csv', sep = '|')

publishers = ['MITOCW']


for publisher in publishers:
    path = path.replace(main_publisher, publisher)
    cid_current = df_main_chapters['Cid'].values.max() + 1
    eid_current = df_main_episodes['Eid'].values.max() + 1 
    sid_current = df_main_series['Sid'].values.max() + 1

    df_second_series = pd.read_csv(path + '/series.csv')
    df_second_series['nSid'] = df_second_series['Sid'] + sid_current

    df_second_episodes = pd.read_csv(path + '/episodes.csv')
    df_second_episodes['nEid'] = df_second_episodes['Eid'] + eid_current
    df_second_episodes['nSid'] = pd.merge(df_second_series, df_second_episodes, on='Sid')['nSid']
    

    df_second_chapters = pd.read_csv(path + '/chapters.csv')
    df_second_chapters['nCid'] = df_second_chapters['Cid'] + cid_current
    df_second_chapters['nEid'] = pd.merge(df_second_episodes, df_second_chapters, on='Eid')['nEid']
    df_second_chapters['nSid'] = pd.merge(df_second_series, df_second_chapters, on='Sid')['nSid']

    df_second_series['Sid'] = df_second_series['nSid']


    df_second_episodes['Sid'] = df_second_episodes['nSid']
    df_second_episodes['Eid'] = df_second_episodes['nEid']
    
    df_second_chapters['Sid'] = df_second_chapters['nSid']
    df_second_chapters['Eid'] = df_second_chapters['nEid']
    df_second_chapters['Cid'] = df_second_chapters['nCid']

    df_main_series = pd.concat([df_main_series, df_second_series], ignore_index = True)
    df_main_episodes = pd.concat([df_main_episodes, df_second_episodes], ignore_index = True)
    df_main_chapters = pd.concat([df_main_chapters, df_second_chapters], ignore_index = True)

    df_main_series.drop(['Author', 'Sessions', 'Material Url', 'Course Structure', 'Episodes', 'nSid'], axis = 1, inplace = True)
    df_main_episodes.drop(['nEid', 'nSid', 'N In Series', 'Course Number', 'Video', 'Videos', 'Date', 'Ressources', 'Chapters', 'Transcriptions'], axis = 1, inplace = True)
    df_main_chapters.drop(['nCid', 'nEid', 'nSid', 'PartN'], axis = 1, inplace = True)

df_main_episodes[1055:1065].head(20)
```

```python
df_main_series.to_csv('../Data/Merged/series.csv', index = False, sep = '|')
df_main_episodes.to_csv('../Data/Merged/episodes.csv', index = False, sep = '|')
df_main_chapters.to_csv('../Data/Merged/chapters.csv', index = False, sep = '|')
```
